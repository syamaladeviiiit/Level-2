```python

filename='Data Files\contacts.txt'
def csvToList(filename):
    li=[]
    with open(filename ,'r') as f:
        for line in f:
            li.append(line.split(','))
    return li
csvToList(filename)

def listToFile(li):
    s=''
    for i in li:
        s += s.join(i)
    return s
li=csvToList(filename)
listToFile(li)
```
